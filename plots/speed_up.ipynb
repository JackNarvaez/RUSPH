```python
import matplotlib.pyplot as plt
import numpy as np
```

# Sedov Wave

```python
thrsd, timsd1, timsd10, timsd100 = np.loadtxt("../Data/speed_up/sedov/scaling_sedov.txt", unpack=1)
T1 = timsd1[0]/timsd1
T10 = timsd10[0]/timsd10
T100 = timsd100[0]/timsd100
```

```python
plt.plot(range(1, 100), range(1, 100),label="Theorical",linestyle="dashed", color="k")
plt.scatter(thrsd, T1, marker='o', color="r", facecolor='none', label="1K Particles")
plt.scatter(thrsd, T10, marker='o', color="b", facecolor='none', label="10K Particles")
plt.scatter(thrsd, T100, marker='o', color="g", facecolor='none', label="100K Particles")

plt.xscale("log")
plt.yscale("log")
plt.xlabel("Ncpus")
plt.ylabel("Speedup")
plt.title(f"Speedup with N=1000 ({int(thrsd[-1])} cores)")
plt.legend()
plt.xlim(1, 100)
plt.xlim(1, 100)
#plt.savefig("speedupsedov.png",dpi=200)
```

# Turbulent Gas

```python
thrtb, timtb32, timtb64 = np.loadtxt("../Data/speed_up/turbulence/scaling_turbulence.txt", unpack=1)
T32 = timtb32[0]/timtb32
T64 = timtb64[0]/timtb64
```

```python
plt.plot(range(1, 100), range(1, 100),label="Theorical",linestyle="dashed", color="k")
plt.scatter(thrtb, T32, marker='o', color="r", facecolor='none', label="32K Particles")
plt.scatter(thrtb[:-1], T64[:-1], marker='o', color="b", facecolor='none', label="262K Particles")
#plt.scatter(thrsd, T100, marker='o', color="g", facecolor='none', label="100K Particles")

plt.xscale("log")
plt.yscale("log")
plt.xlabel("Ncpus")
plt.ylabel("Speedup")
plt.title(f"Speedup with N=1000 ({int(thrsd[-1])} cores)")
plt.legend()
plt.xlim(1, 100)
plt.xlim(1, 100)
#plt.savefig("speedupturbulence.png",dpi=200)
```
