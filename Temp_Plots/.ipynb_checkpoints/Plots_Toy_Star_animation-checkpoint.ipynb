```python
# Angular Momentum
def angular_momentum(x, y, vx, vy):
    return x*vy - y*vx
```

```python
T = 73
time = range(0, T)
ang_mom = np.zeros(T)

fig = plt.figure()

for ii in range(1, T+1):
    df = src.read_csv('../Data/results/toy_star/' + str(ii*100) + ".csv")
    df.params={'mass': 2.0/df.shape[0]}
    ax1 = fig.add_subplot(1,1,1)
    plot_df = df.render('rho', ax= ax1, xsec=1.0)
    ax1.set_xlim(-1, 1)
    ax1.set_ylim(-1, 1)
    fig.savefig("../Data/results/toy_star/images/image_" + str(ii) + ".png")
    ax1.cla()
    plt.clf()
    ang_mom[ii-1] = np.sum(angular_momentum(df['x'], df['y'], df['vx'], df['vy']))
plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
plt.plot(time, ang_mom, '-', color='crimson')
plt.show()
```
