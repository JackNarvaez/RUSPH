```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
threads, time = np.loadtxt("../Data/speed_up/scaling.txt", unpack=1)
T = time[0]/time
```

```python
plt.plot(threads, threads,label="Theorical",linestyle="dashed", color="crimson")
plt.scatter(threads, T, marker='+', color="k", label="Experimental")

plt.xscale("log")
plt.yscale("log")
plt.xlabel("Ncpus")
plt.ylabel("Speedup")
plt.title(f"Speedup with N=1000 ({threads.size//2} cores/{threads.size} threads)")
plt.legend()
#plt.xlim(10, 40)
#plt.xlim(10, 40)
#plt.savefig("speedup.png",dpi=200)
```
